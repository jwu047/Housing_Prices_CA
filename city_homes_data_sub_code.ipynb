```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
city_homes_data = pd.read_csv("City_Homes.csv")
city_homes_data["Average Median Home Value"] = (city_homes_data["2016-01"]+city_homes_data["2016-02"]+city_homes_data["2016-03"]+city_homes_data["2016-04"]+
                              city_homes_data["2016-05"]+city_homes_data["2016-06"]+city_homes_data["2016-07"]+city_homes_data["2016-08"]+
                              city_homes_data["2016-09"]+city_homes_data["2016-10"]+city_homes_data["2016-11"]+city_homes_data["2016-12"])/12
city_homes_data = city_homes_data[["City", "State", "CountyName", "Average Median Home Value"]]
#city_homes_data.head()

# create new data frame which looks for minority in california only
city_homes_california=city_homes_data.loc[city_homes_data["State"]=="CA"]
city_homes_california
```

```python
counties=city_homes_california.groupby("CountyName")
city_homes_counties = counties["Average Median Home Value"].mean()
print(city_homes_counties)
```

```python
#Med_homes.dtypes

# to data frame table
city_homes_new=pd.DataFrame(city_homes_counties)
city_homes_new

#save
city_homes_new.to_csv("city_homes_new.csv", encoding="utf-8", index=True)
```
