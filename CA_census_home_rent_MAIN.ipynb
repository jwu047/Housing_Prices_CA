```python
# Import dependencies

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
```

```python
# set path for file
census_file="census_data_updatePD.csv"
census_pd = pd.read_csv(census_file)
census_pd.head()
```

```python
# plots()
plt.scatter(census_pd["Poverty Rate"], census_pd["Median Home Value"], marker="o", facecolors="red")
plt.ylim(0, 2100000)
plt.title("Median home value vs Poverty rate in California")
plt.xlabel("Poverty rate (%)")
plt.ylabel("Median home value ($)")
plt.savefig("Median_home_value_vs_PovertyRate in California.png")
```

```python
# Plot (Median home value vs city)
plt.scatter(census_pd["Zipcode"], census_pd["Median Home Value"], marker = "+", facecolors="blue")
plt.title("Median home value vs. Zip code in CA")
plt.xlabel("Zip code")
plt.ylabel("Median home value ($)")
plt.savefig("Median_home_value_vs_zip codes in California.png")
```

```python
# plot household income vs Median home values
plt.scatter(census_pd["Household Income"], census_pd["Median Home Value"], marker="_", facecolors="skyblue")
plt.title("Median home value vs. Household income in California")
plt.ylabel("Median home value ($)")
plt.xlabel("House hole income ($)")
plt.savefig("Median_home_value_vs_Household income in CA.png")
```

```python
# PLOT RENT PER ZIP CODES

plt.scatter(census_pd["Zipcode"], census_pd["average rent"], marker="_", facecolors="skyblue")
plt.title("Average rent vs. zip code in California")
plt.ylabel("Average rent ($)")
plt.xlabel("Zip code in California")
plt.savefig("Average rent vs zip codes in california.png")
```

```python
# Plot rent vs home values

plt.scatter(census_pd["Median Home Value"], census_pd["average rent"], marker="o", facecolors="blue")
plt.title("Average rent vs. Median Home Value in California")
plt.ylabel("Average rent ($)")
plt.xlabel("Median home value in California ($)")
plt.savefig("Average rent vs home value in california.png")
plt.show()
```

```python
# PLOT MEDIAN HOME VALUE PER ZIP CODES

plt.scatter(census_pd["Zipcode"], census_pd["Median Home Value"], marker="_", facecolors="blue")
plt.title("Median home value vs. zip code in California")
plt.ylabel("Median home value ($)")
plt.xlabel("Zip code in California")
plt.savefig("Median home value_vs zip codes in california.png")
```

```python
# PLOT HOUSEHOLD INCOME PER ZIP CODES

plt.scatter(census_pd["Zipcode"], census_pd["Household Income"], marker="_", facecolors="blue")
plt.title("Household income vs. zip code in California")
plt.ylabel("Household income ($)")
plt.xlabel("Zip code in California")
plt.savefig("Household income vs zip code in california.png")
```

```python
# RUN STATISTIC
%matplotlib inline
import scipy.stats as stats

y1=census_pd["Zipcode"]
y2=census_pd["average rent"]
y3=census_pd["Median Home Value"]
y4=census_pd["Household Income"]

p1=stats.pearsonr(y2,y3)[1]
p2=stats.pearsonr(y3,y4)[1]
print(f"p-value for Avg rent & Median Home value: {p1}, --> correlate")
print(f"p-value for Home value & income: {p2}, --> correlate")
```

```python
# Poverty & home value
y5=census_pd["Poverty Rate"]
p3 = stats.pearsonr(y5, y3)[1]
print(f"p-value for poverty rate and home value: {p3}, --> correlate")
```
