```python
import pandas as pd
```

```python
# Read Zillow Home Value Data
zillow_df=pd.read_csv("Zip_Zhvi_AllHomes.csv")
zillow_df.head()
```

```python
#filter california dta only
zillow_ca_df = zillow_df.loc[zillow_df["State"]=="CA",:]
zillow_ca_df = zillow_ca_df.loc[:,["RegionName","City","State","CountyName","2016-01","2016-02","2016-03","2016-04","2016-05","2016-06","2016-07","2016-08","2016-09","2016-10","2016-11","2016-12"]]
zillow_ca_df = zillow_ca_df.rename(columns={"RegionName":"Zipcode"})
zillow_ca_df.head()
```

```python
# Save as a csv
zillow_ca_df.to_csv("zillow_ca_homevalue.csv", index=False, header = True)
```
